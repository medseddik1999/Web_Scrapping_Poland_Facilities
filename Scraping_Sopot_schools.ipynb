```python
import requests 
from bs4 import BeautifulSoup
import re # regex expression
from tqdm import tqdm # time loop in notebook 
import translators as ts   
from time import sleep  
import pandas as pd  
#from itertools import chain
```

```python
#PRIMARY SCHOOLS (0-10)
response=requests.get('https://katalog.trojmiasto.pl/nauka_i_szkolnictwo/szkoly_podstawowe/sopot/#s...@sp1sopot.eu')
content=response.content
soup=BeautifulSoup(content, 'html.parser')

list_name_non_clean=soup.find_all(class_='objectName')
list_name_non_clean

#NAMES 

def clean_name(x):
    t=re.search('<span>(.*)</span></a>', x).group(1)
    return t

list_name_clean=[]
for item in list_name_non_clean:
    nn=clean_name(str(item))
    list_name_clean.append(nn)
```

```python
colonne = ['Facility.Name','Primary school', 'Secondary school','Facility.Tel_Number (+48)','Facility.Web_site','Facility.Address','Facility.email','Facility.Courts','Facility.Sport_activities']
df= pd.DataFrame(columns=colonne)
df['Facility.Name']=list_name_clean
df['Primary school']=1
df['Secondary school']=0
df
```

```python
list_pho_non_clean=soup.find_all(class_='objectPhone')

def clean_pho(x):
    t=re.search('"tel\:(.*)">', x).group(1)
    return t

list_pho_clean=[]
for item in list_pho_non_clean:
    nn=clean_pho(str(item))
    list_pho_clean.append(nn)
    
list_pho_clean[0]='606610102'
list_pho_clean[1]='515256108'
list_pho_clean[4]='585513824'
list_pho_clean[7]='881313343'
list_pho_clean

df['Facility.Tel_Number (+48)']=list_pho_clean
df
```

```python
list_web_clean=['www.lokomotywa.edu.pl','sopotmontessori.com','www.ssa.gfo.pl','www.sopockaszkolamuzyczna.pl,','www.sp1sopot.eu','sp9.edu.pl','zss5sopot.pl','sp8sopot.edupage.org','www.sp7sopot.szkolna.net,','pcrsopot.pl/abc-pacjenta/szkola-przyszpitalna/','latawiecmontessori.com']
df['Facility.Web_site']=list_web_clean
```

```python
list_add_non_clean=soup.find_all(class_='objectAddress__street')
def clean_add(x):
    t=re.search('<span class="objectAddress__street">(.*)</span>', x).group(1)
    return t

list_add_clean=[]
for item in list_add_non_clean:
    nn=clean_add(str(item))
    list_add_clean.append(nn)
    
list_cap_non_clean=soup.find_all(class_='objectAddress__postCode')
def clean_cap(x):
    t=re.search('<span class="objectAddress__postCode">(.*)</span>', x).group(1)
    return t

list_cap_clean=[]
for item in list_cap_non_clean:
    nn=clean_cap(str(item))
    list_cap_clean.append(nn)
    
list_cap_clean

address=[list_add_clean[i] + ', '+list_cap_clean[i]+ ', Sopot' for i in range(11)]
df['Facility.Address']=address
df
```
