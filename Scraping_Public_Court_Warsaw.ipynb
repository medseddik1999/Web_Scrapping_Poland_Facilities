```python
import requests 
from bs4 import BeautifulSoup
import re # regex expression
from tqdm import tqdm # time loop in notebook 
import re  
import translators as ts   
from time import sleep  
import pandas as pd  
from itertools import chain
```

```python
#i=list(range(1,108))  
def Clean (x): 
    t=re.search('</span>(.*)</td>' ,x).group(1)  
    return t  

def Clean2 (x): 
    t=re.search('<span class="heading-label">(.*)</span>' ,x).group(1)  
    return t  


def translate(x):  
    if x ==' ' : 
        t=' ' 
    else : 
        t=ts.google(x, from_language='pl', to_language='en')   
    return t 
```

```python
data_list=[] 
for i in tqdm(range(1,108)): 
    sleep(5)
    url="https://edukacja.um.warszawa.pl/boiska-szkolne-otwarte-dla-mieszkancow?p_p_id=com_liferay_asset_publisher_web_portlet_AssetPublisherPortlet_INSTANCE_PyBRvBqQE6qx&p_p_lifecycle=0&p_p_state=normal&p_p_mode=view&_com_liferay_asset_publisher_web_portlet_AssetPublisherPortlet_INSTANCE_PyBRvBqQE6qx_delta=10&p_r_p_resetCur=false&_com_liferay_asset_publisher_web_portlet_AssetPublisherPortlet_INSTANCE_PyBRvBqQE6qx_cur="+str(i)
    response = requests.get(url)
    result = response.content

    # Parse html with BS
    soup = BeautifulSoup(result, 'html.parser')  
    content=soup.find_all('td')  
    content=list(map(str ,content)) 
    final_content=[]
    for j in content :
        chart=" ".join(j.split()) 
        final_content.append(chart)  
    elements = list(map(Clean,final_content)) 
    data_list.append(elements) 
```

```python
data_list=list(chain.from_iterable(data_list)) #delet sublistes 
```

```python
chunks = [data_list[x:x+18] for x in range(0, len(data_list), 18)]  
```

```python
Heading_label_list = soup.find_all('span' ,class_="heading-label" )   
Heading_label_list =list(map(str ,Heading_label_list)) 
Heading_label_list=list(map(Clean2,Heading_label_list))  
Heading_label_list=list(map(translate ,Heading_label_list)) 

```

```python
df= pd.DataFrame(chunks, columns = Heading_label_list)   
```

```python
days=["Monday","Tuesday","Wednesday","Thursday"	,"Friday" ,"Saturday","Sunday"] 
taktak=["Does the object have a sanitary interchange (toilet)?","Is the object illuminated?","Is the facility equipped with stands?"]
```

```python
def tra_time(x):
    t=x.replace('od','from') 
    t=t.replace('do','to') 
    return t  

def TAk_NIE(x): 
    t=x.replace('TAK','YES') 
    t=t.replace('NIE','NO') 
    return t 
```

```python
for day in days : 
    df[day]=list(map(str,df[day])) 
    df[day]=list(map(tra_time,df[day])) 
```

```python
for tak in taktak : 
    df[tak]=list(map(str,df[tak])) 
    df[tak]=list(map(TAk_NIE,df[tak])) 
```

```python
df['type of object']=df['type of object'].map(lambda t : t.replace('jasts','jumps'))  
```
